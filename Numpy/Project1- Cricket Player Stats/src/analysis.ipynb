```python
import numpy as np
import csv
```

Import CSV into NumPy

```python

players = []
stats = []

with open("../data/players_stats.csv", "r") as f:
    reader = csv.reader(f)
    next(reader)  # skip header
    for row in reader:
        players.append(row[0])
        stats.append(list(map(float, row[1:])))

players = np.array(players)
stats = np.array(stats)

print("Loaded players:", players.shape)
print("Loaded stats:", stats.shape)
```

Basic Statistics

```python
mean_stats = stats.mean(axis=0)
max_sr_player = players[np.argmax(stats[:, 1])]
max_wickets_player = players[np.argmax(stats[:, 3])]

print("\nMean Stats:", mean_stats)
print("Best Strike Rate:", max_sr_player)
print("Most Wickets:", max_wickets_player)
```

Role Classification

```python
batsmen = players[stats[:, 0] > 5000]
bowlers = players[stats[:, 3] > 100]
allrounders = players[(stats[:, 0] > 5000) & (stats[:, 3] > 100)]

print("\nBatsmen:", batsmen)
print("Bowlers:", bowlers)
print("All-Rounders:", allrounders)
```

Performance Score Ranking

```python
score = (
    stats[:, 0] * 0.002 +
    stats[:, 1] * 0.8 +
    stats[:, 2] * 1.5 +
    stats[:, 3] * 0.5 -
    stats[:, 4] * 1.2
)

ranking_indices = np.argsort(score)[::-1]
top_3 = players[ranking_indices[:3]]

print("\nTop 3 Players:", top_3)
```

Export Scored Data to CSV

```python
output_file = "../data/players_scores.csv"

with open(output_file, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Player", "Score"])
    for name, s in zip(players, score):
        writer.writerow([name, round(s, 2)])

print("\nExported scored data to:", output_file)
```
